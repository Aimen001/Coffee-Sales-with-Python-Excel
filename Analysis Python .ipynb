```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dirty_data_cafe_sales.csv')
```

```python
df.head()
```

```python
### check the null values
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
```

```python
### Replace the Unknown with the mode
columns_to_replace = [
    'Item', 'Quantity', 'Price Per Unit', 'Total Spent',
    'Payment Method', 'Location', 'Transaction Date'
]

for col in columns_to_replace:
    if col in df.columns:
        # Replace 'UNKNOWN' with the mode of the column
        mode_value = df[col].mode()[0]
        df[col] = df[col].replace('UNKNOWN', mode_value)
```

```python
df.head(15)
```

```python
### Replace the Error with the mode
columns_to_replace = [
    'Item', 'Quantity', 'Price Per Unit', 'Total Spent',
    'Payment Method', 'Location', 'Transaction Date'
]

for col in columns_to_replace:
    if col in df.columns:
        # Replace 'Error' with the mode of the column
        mode_value = df[col].mode()[0]
        df[col] = df[col].replace('ERROR', mode_value)
```

```python
df.head(15)
```

```python
### filling the null values
columns_to_replace = [
    'Item', 'Quantity', 'Price Per Unit', 'Total Spent',
    'Payment Method', 'Location', 'Transaction Date'
]

for col in columns_to_replace:
    if col in df.columns:  # Ensure the column exists
        mode_value = df[col].mode()[0]  # Get the mode of the column
        df[col] = df[col].fillna(mode_value)
```

```python
df.isnull().sum()
```

```python
###changing the transaction date to daytime
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```

```python
df['Transaction Date'] = df['Transaction Date'].fillna(df['Transaction Date'].mode()[0])
```

```python
### split the day time
df['Year'] = df['Transaction Date'].dt.year
df['Month'] = df['Transaction Date'].dt.month
df['Day'] = df['Transaction Date'].dt.day
df['Day Name'] = df['Transaction Date'].dt.day_name()
```

```python
df.dtypes
```

```python
### changing the data type to numeric
columns_to_convert = ['Quantity', 'Price Per Unit', 'Total Spent']

df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
columns_to_convert = ['Year', 'Month', 'Day']

df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
### checking if is there any null values left
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df.dtypes
```

```python
### calcutale total sales
df['Total_sales'] = df['Quantity'] * df['Price Per Unit']
print(df['Total_sales'].sum())
```

```python
### what items are selling the most and why?

best_sales_item = df.groupby('Item')['Transaction ID'].count().reset_index()
sns.barplot(x = 'Item', y = 'Transaction ID', data = best_sales_item,  palette = 'viridis' )
plt.title('best_sales_item', fontsize=16)
plt.xlabel('Item', fontsize=14)
plt.ylabel('Number of transactions', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
### What months sell the most and why?
sales_per_month = df.groupby('Month')['Total_sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Total_sales', data=sales_per_month, palette='viridis')
plt.title('Total Sales Per Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
### What method does the client like to use and does it 
fav_method = df.groupby('Payment Method')['Transaction ID'].count().reset_index()
sns.barplot(x = 'Payment Method', y = 'Transaction ID', data = fav_method, palette='viridis')
plt.title('Favorite Payment Method', fontsize=16)
plt.xlabel('Payment Method', fontsize=14)
plt.ylabel('Number of Transactions', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
### Whats the prefer location for our clients is it take-way or In-store?
fav_location = df.groupby('Location')['Transaction ID'].count().reset_index()

sns.barplot(x = 'Location', y = 'Transaction ID', data = fav_location, palette='viridis')
plt.title('Favorite Location', fontsize=16)
plt.xlabel('Location', fontsize=14)
plt.ylabel('Number of Transactions', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
### Does location have any relation with the payment method?
plt.figure(figsize=(12, 6))
sns.countplot(x='Location', hue='Payment Method', data=df, palette='viridis')
plt.title('Distribution of Payment Methods by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
### Which day should the bussiness sell at
best_day = df.groupby('Day Name')['Transaction ID'].count().reset_index()

sns.barplot(x = 'Day Name', y = 'Transaction ID', data = best_day, palette='viridis')
plt.title('Best day to sell at', fontsize=16)
plt.xlabel('Days', fontsize=14)
plt.ylabel('Number of Transactions', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
